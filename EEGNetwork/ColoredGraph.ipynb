```python
import matplotlib.pyplot as plt
import numpy as np
import glob
import shutil
import cv2
```

```python
#Import the documents from Google Drive and extract filename instead of the full path
#Change path name accordingly to where the file is located.
sub = sorted(glob.glob("/content/drive/My Drive/Spark-2019/RECSEM2019/EEG Network/Sub-cof-?.txt")) + sorted(glob.glob("/content/drive/My Drive/Spark-2019/RECSEM2019/EEG Network/Sub-cof-??.txt"))
sub = [x[58:-4] for x in sub]
print(sub)

```

```python
#Loop through all of the correlation matrices for each subject and
#Plot them in gray-scale to visualize the causal realtionship between each pair of channels
for s in sub:
  with open("/content/drive/My Drive/Spark-2019/RECSEM2019/EEG Network/" + s +".txt", 'r') as f:
    lines = f.readlines()[2:]
    for i in range(len(lines)):
        lines[i] = [float(x) for x in lines[i].split('|')[1:-1]]

    matrix = np.asarray(lines)
    plt.figure()
    plt.imshow(matrix, cmap='gray');
    #plt.colorbar(extend='both')
    plt.axis('off')
    fig1 = plt.gcf()
    plt.show()
    fig1.savefig("/content/drive/My Drive/Spark-2019/RECSEM2019/EEG Network/"+s+'-gray.png',bbox_inches='tight')
```

```python
m = sorted(glob.glob('/content/drive/My Drive/Spark-2019/RECSEM2019/EEG Network/Threshold(0.7)/Threshold-?.png'))+sorted(glob.glob('/content/drive/My Drive/Spark-2019/RECSEM2019/EEG Network/Threshold(0.7)/Threshold-??.png'))

#Import newly created gray-scale maps to resize the image into a symmetric shape
#Resizing helps to ensure that all the images will have the same size
#for Neural Network training in MATLAB
for path in m:
  img = cv2.imread(path, cv2.IMREAD_UNCHANGED)
 
  print('Original Dimensions : ',img.shape)

  width = 250
  height = 250 # keep original height
  dim = (width, height)

  # resize image
  resized = cv2.resize(img, dim, interpolation = cv2.INTER_AREA)
  cv2.imwrite(path, resized)

  print('Resized Dimensions : ',resized.shape)
  print("---------")
```
